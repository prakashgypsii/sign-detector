```python
from google.colab import files
uploaded = files.upload()

# Step 2: Load the Trained YOLOv8 Model
from ultralytics import YOLO
model = YOLO("/content/runs/detect/train3/weights/best.pt")  # adjust path if needed

# Step 3: Run Detection on Uploaded Image
image_path = list(uploaded.keys())[0]
results = model.predict(source=image_path, save=True, imgsz=640)

# Step 4: Display Annotated Output Image
from IPython.display import Image, display
import os

predict_folder = "runs/detect/predict"
result_image_path = os.path.join(predict_folder, os.listdir(predict_folder)[0])
display(Image(filename=result_image_path))

# Step 5: Crop and Display Detected Signature
import cv2
from PIL import Image as PILImage

img = cv2.imread(result_image_path)

for r in results:
    for box in r.boxes.xyxy:
        x1, y1, x2, y2 = map(int, box)
        cropped = img[y1:y2, x1:x2]
        cv2.imwrite("cropped_signature.jpg", cropped)

# Show the cropped signature
display(PILImage.open("cropped_signature.jpg"))
```
